# 12. Loop para diferentes SMA

```python
def SMA_dataset(ticker, short, long, period_init, period_final):
    yf.pdr_override()
    data = web.get_data_yahoo(tickers, start=start_date, end=end_date)
    data_adj_close = data.loc[:,"Adj Close"].round(2)

    data_moving_average = pd.DataFrame()
    data_moving_average['BBAS3.SA'] = data_adj_close[tickers[0]]
    data_moving_average['BBAS3.SA_Short'] = data_adj_close[tickers[0]].rolling(window=short).mean()
    data_moving_average['BBAS3.SA_Long'] = data_adj_close[tickers[0]].rolling(window=long).mean()
    
    data_moving_average['ITUB4.SA'] = data_adj_close[tickers[0]]
    data_moving_average['ITUB4.SA_Short'] = data_adj_close[tickers[1]].rolling(window=short).mean()
    data_moving_average['ITUB4.SA_Long'] = data_adj_close[tickers[1]].rolling(window=long).mean()
    
    data_moving_average['SANB11.SA'] = data_adj_close[tickers[0]]
    data_moving_average['SANB11.SA_Short'] = data_adj_close[tickers[2]].rolling(window=short).mean()
    data_moving_average['SANB11.SA_Long'] = data_adj_close[tickers[2]].rolling(window=long).mean()
    
    data_moving_average['ABCB4.SA'] = data_adj_close[tickers[0]]
    data_moving_average['ABCB4.SA_Short'] = data_adj_close[tickers[3]].rolling(window=short).mean()
    data_moving_average['ABCB4.SA_Long'] = data_adj_close[tickers[3]].rolling(window=long).mean()
    
    data_moving_average['BBDC4.SA'] = data_adj_close[tickers[0]]
    data_moving_average['BBDC4.SA_Short'] = data_adj_close[tickers[4]].rolling(window=short).mean()
    data_moving_average['BBDC4.SA_Long'] = data_adj_close[tickers[4]].rolling(window=long).mean()
    
    data_moving_average = data_moving_average.dropna()
    return data_moving_average

def sma_graphs(data_moving_average):
    fig, ax = plt.subplots(5,1, figsize=(15,15))
    fig.subplots_adjust(hspace=0.5, wspace=0.3)
    lw=3 
    color_short='g'
    color_long = 'r'
    linestyle='--'
    
    ax[0].plot(data_moving_average['BBAS3.SA_Short'], label='BBAS3.SA_Short', color=color_short)
    ax[0].plot(data_moving_average['BBAS3.SA_Long'], label='BBAS3.SA_Long', color=color_long)
    ax[0].set_title(f"{tickers[0]} moving average")
    
    ax[1].plot(data_moving_average['ITUB4.SA_Short'], label='ITUB4.SA_Short', color=color_short)
    ax[1].plot(data_moving_average['ITUB4.SA_Long'], label='ITUB4.SA_Long', color=color_long)
    ax[1].set_title(f"{tickers[1]} moving average")
    
    
    ax[2].plot(data_moving_average['SANB11.SA_Short'], label='SANB11.SA_Short', color=color_short)
    ax[2].plot(data_moving_average['SANB11.SA_Long'], label='SANB11.SA_Long', color=color_long)
    ax[2].set_title(f"{tickers[2]} moving average")
    
    
    ax[3].plot(data_moving_average['ABCB4.SA_Short'], label='ABCB4.SA_Short', color=color_short)
    ax[3].plot(data_moving_average['ABCB4.SA_Long'], label='ABCB4.SA_Long', color=color_long)
    ax[3].set_title(f"{tickers[3]} moving average")
    
    
    ax[4].plot(data_moving_average['BBDC4.SA_Short'], label='BBDC4.SA_Short', color=color_short)
    ax[4].plot(data_moving_average['BBDC4.SA_Long'], label='BBDC4.SA_Long', color=color_long)
    ax[4].set_title(f"{tickers[4]} moving average")
    
    
    for i in range (4):
        ax[i].set_ylabel("Price")
    plt.show()
```

```python
def sma_movements(ticker, data_moving_average):
    ticker_short = ticker + "_Short"
    ticker_long = ticker  + "_Long"
    data_moving_average_ticker = data_moving_average[[ticker, ticker_short, ticker_long]]
    data_moving_average_ticker['Movement'] = np.where(data_moving_average_ticker[ticker_short] > data_moving_average_ticker[ticker_long], 1, -1)
    return data_moving_average_ticker

def sma_graph_movements(ticker, data_moving_average):
    ticker_short = ticker + "_Short"
    ticker_long = ticker  + "_Long"
    fig, ax = plt.subplots(1,1, figsize=(15,6))
    fig.subplots_adjust(hspace=0.5, wspace=0.3)
    lw=3 
    color_short='g'
    color_long = 'r'
    linestyle='--'
    
    ax.plot(data_moving_average_ticker[ticker_short], label=ticker_short, color=color_short)
    ax.plot(data_moving_average_ticker[ticker_long], label=ticker_long, color=color_long)
    ax.set_title(f"{tickers[0]} moving average")
    
    for i, ret in enumerate(data_moving_average_ticker['Movement']):
        if ret == 1:
            ax.fill_between(data_moving_average_ticker.index[i:i+2],
                            data_moving_average_ticker[ticker_short].min(),
                            data_moving_average_ticker[ticker_long].max(),
                            color='green', alpha=0.3)
        elif ret == -1:
            ax.fill_between(data_moving_average_ticker.index[i:i+2],
                            data_moving_average_ticker[ticker_short].min(),
                            data_moving_average_ticker[ticker_long].max(),
                            color='red', alpha=0.3)
```

```python
def sma_returns(ticker, data_moving_average_ticker):
    data_moving_average_ticker_buy = data_moving_average_ticker[data_moving_average_ticker['Movement'] == 1]
    data_moving_average_ticker_buy['ma_return'] = data_moving_average_ticker_buy[ticker].pct_change()
    data_moving_average_ticker_buy['cumm_return'] = data_moving_average_ticker_buy.ma_return.cumsum().apply(np.exp)

def sma_returns_graph(ticker, data_moving_average_ticker):
    fig, ax = plt.subplots(1,1, figsize=(15,6))
    fig.subplots_adjust(hspace=0.5, wspace=0.3)
    lw=3 
    color_graph = 'g'
    color_line ='r'
    linestyle='--'
    ponto = (data_moving_average_ticker_buy.ma_return.sum()).round(2)
    ax.plot(data_moving_average_ticker_buy['cumm_return'], label=f'{ticker} SMA Cummulative Return', color=color_graph)
    plt.axhline(y=np.exp(data_moving_average_ticker_buy.ma_return.sum()).round(2), color=color_line, linestyle=linestyle)
    plt.show()
```

```python
def drawdowns(ticker, data_moving_average_ticker, show_graphs):
    data_moving_average_ticker_buy['cummax'] = data_moving_average_ticker_buy.cumm_return.cummax()
    data_moving_average_ticker_buy

    data_moving_average_ticker_buy['drawdown %'] = (data_moving_average_ticker_buy['cummax'] - data_moving_average_ticker_buy['cumm_return'])/data_moving_average_ticker_buy['cummax']
    data_moving_average_ticker_buy.round(2)
    
    if show_graphs == True:
        fig, ax = plt.subplots(1,1, figsize=(15,6))
        fig.subplots_adjust(hspace=0.5, wspace=0.3)
        lw=3 
        color_graph = 'g'
        color_line ='r'
        linestyle='--'
        ponto = (data_moving_average_ticker_buy.ma_return.sum()).round(2)
        ax.plot(data_moving_average_ticker_buy['cummax'], label='BBAS3 Cummulative Return', color='b')
        ax.plot(data_moving_average_ticker_buy['cumm_return'], label='BBAS3 Cummulative Return', color='g')
        
        plt.axhline(y=np.exp(data_moving_average_ticker_buy.ma_return.sum()).round(2), color=color_line, linestyle=linestyle)
        plt.show()
```
